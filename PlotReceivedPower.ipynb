## Parse CSV File


```
from google.colab import drive
drive.mount("/content/gdrive")
csv_path = "gdrive/My Drive/UofT/4th Year/capstone_repos/SimulatorInterface/Sample/ReceivedPowers.csv"
```

```
import pandas as pd
received_powers_df = pd.read_csv(csv_path, sep=',', skipinitialspace=True)
x_positions = received_powers_df["X[m]"].tolist()
y_positions = received_powers_df["Y[m]"].tolist()
z_positions = received_powers_df["Z[m]"].tolist()
received_powers = received_powers_df["Power[dBm]"].tolist()
```

##Plots

```
import matplotlib.pyplot as plt
```

### 3D plot - X, Y, Receiver Power

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(x_positions, y_positions, received_powers)
ax.set_xlabel('X[m]')
ax.set_ylabel('Y[m]')
ax.set_zlabel('Received Power[dBm]')
plt.tight_layout()
```

### 3D plot - X, Y, Z, Received Power (power is the colour)



```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(x_positions, y_positions, z_positions, c=received_powers, cmap=plt.get_cmap('seismic'))
ax.set_xlabel('X[m]')
ax.set_ylabel('Y[m]')
ax.set_zlabel('Z[m]')
plt.tight_layout()
```
